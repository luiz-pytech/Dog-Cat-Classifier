```python
## Importando as libs necessrias
from google.colab import files
import os, io, time

# Criando o diretrio inicial: /content/
os.chdir('/content/')
try:
  os.mkdir('Imagens', )
except:
  print('A pasta j existe.')
os.chdir('./Imagens')
os.listdir()
uploaded_images = files.upload()
```

```python
## Reduzir as imagens
from skimage.io import imread, imshow
from skimage.transform import resize
from skimage.feature import hog
from skimage.color import rgb2gray

def aplicar_HOG(tamanho_imagem, pixel_por_celula):
  hog_features = []

  for filename in uploaded_images.keys():
    image = imread(filename)
    image_resized = resize(image, (tamanho_imagem, tamanho_imagem))

    # Se tiver 4 canais (RGBA), converte para RGB
    if image_resized.ndim == 3 and image_resized.shape[2] == 4:
        image_resized = image_resized[:, :, :3]

    # Se tiver 3 canais (RGB), converte para escala de cinza
    if image_resized.ndim == 3 and image_resized.shape[2] == 3:
        gray = rgb2gray(image_resized)
    else:
        # J est em escala de cinza
        gray = image_resized

    fd, hog_image = hog(gray, orientations=9, pixels_per_cell=(pixel_por_celula, pixel_por_celula),
                        cells_per_block=(2, 2), visualize=True)
    hog_features.append(fd)

  return hog_features
```

```python
## Pegando as classes=de cada imagem lida
def extrair_classe():
  classe = []

  for filename in uploaded_images.keys():
    if 'Egyptian_Mau' in filename:
        classe.append('Egyptian_Mau')
    elif 'pomeranian' in filename:
        classe.append('Pomeranian')
    elif 'Ragdoll' in filename:
        classe.append('Ragdoll')
    elif 'shiba_inu' in filename:
        classe.append('Shiba_inu')

  return classe
```

```python
import pandas as pd

def salvar_hog_csv(hog_features, classe, nome_arquivo):
    # Criao do DataFrame
    data = {"hog_features": hog_features, "classe": classe}
    df = pd.DataFrame(data)

    # Expandindo os vetores HOG
    df2 = pd.DataFrame(df['hog_features'].tolist())
    df2.columns = df2.columns.map(lambda x: f'hog_feature_{x+1}')

    # Concatenando e organizando
    df = pd.concat([df.drop('hog_features', axis=1), df2], axis=1)
    df['classe'] = df.pop('classe')

    # Salvando CSV
    df.to_csv(nome_arquivo, sep=',', index=False)
    print(f"Arquivo salvo como {nome_arquivo}")
```

```python
classe = extrair_classe()

hog_features_128_16x16 = aplicar_HOG(128, 16)

print(f"HOG features extrados: {len(hog_features_128_16x16)}")
print(f"Classes extradas: {len(classe)}")
salvar_hog_csv(hog_features_128_16x16, classe, 'HOG_128_16x16.csv')

hog_features_256_16x16 = aplicar_HOG(256, 16)
salvar_hog_csv(hog_features_256_16x16 , classe, 'HOG_256_16x16.csv')

hog_features_256_20x20 = aplicar_HOG(256, 20)
salvar_hog_csv(hog_features_256_20x20, classe, 'HOG_256_20x20.csv')
```
