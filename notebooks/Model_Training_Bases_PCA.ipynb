```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
```

```python
## Bases da dados
df1 = pd.read_csv('../Dados/PCA_Features/PCA_cnn_VGG16_256_max.csv')
df2 = pd.read_csv('../Dados/PCA_Features/PCA_cnn_VGG16_256_avg.csv')
df3 = pd.read_csv('../Dados/PCA_Features/PCA_cnn_VGG19_128_max.csv')
df4 = pd.read_csv('../Dados/PCA_Features/PCA_cnn_VGG19_256_max.csv')
df5 = pd.read_csv('../Dados/PCA_Features/PCA_cnn_VGG19_128_avg.csv')
df6 = pd.read_csv('../Dados/PCA_Features/PCA_cnn_VGG19_256_avg.csv')

def preparation_holdout(dataframe):
    df_features = dataframe.iloc[:, 1:-1]
    df_classe = dataframe['classe']

    return df_features, df_classe;

X1, y1 = preparation_holdout(df1)
X2, y2 = preparation_holdout(df2)
X3, y3 = preparation_holdout(df3)
X4, y4 = preparation_holdout(df4)
X5, y5 = preparation_holdout(df5)
X6, y6 = preparation_holdout(df6)
```

```python
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.3,shuffle=True, random_state=1)
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.3,shuffle=True, random_state=1)
X3_train, X3_test, y3_train, y3_test = train_test_split(X3, y3, test_size=0.3,shuffle=True, random_state=1)
X4_train, X4_test, y4_train, y4_test = train_test_split(X4, y4, test_size=0.3,shuffle=True, random_state=1)
X5_train, X5_test, y5_train, y5_test = train_test_split(X5, y5, test_size=0.3,shuffle=True, random_state=1)
X6_train, X6_test, y6_train, y6_test = train_test_split(X6, y6, test_size=0.3,shuffle=True, random_state=1)
```

```python
def aplicar_CNN(X_train, X_test, y_train, y_test, k_range = 10):
    for i in range(1, k_range + 1):
        kNN = KNeighborsClassifier(n_neighbors=i, metric='euclidean') 
        kNN.fit(X_train, y_train)
    
        y_pred = kNN.predict(X_test)
        print(f"Accuracy k = {i}:",metrics.accuracy_score(y_test, y_pred))
        print(confusion_matrix(y_test, y_pred))

print("\nPCA_cnn_VGG16_256_max:")
aplicar_CNN(X1_train, X1_test, y1_train, y1_test)

print("\nPCA_cnn_VGG16_256_avg:")
aplicar_CNN(X2_train, X2_test, y2_train, y2_test)

print("\nPCA_cnn_VGG19_128_max:")
aplicar_CNN(X3_train, X3_test, y3_train, y3_test)

print("\nPCA_cnn_VGG19_256_max:")
aplicar_CNN(X4_train, X4_test, y4_train, y4_test)

print("\nPCA_cnn_VGG19_128_avg")
aplicar_CNN(X5_train, X5_test, y5_train, y5_test)

print("\nPCA_cnn_VGG19_256_avg:")
aplicar_CNN(X6_train, X6_test, y6_train, y6_test)

```

```python
def aplicar_KNN_with_kfold(X, y, k_range = 10, k_fold = 10):
    kf = KFold(n_splits= k_fold, random_state=1, shuffle=True)
    for i in range(1, k_range + 1):
        kNN = KNeighborsClassifier(n_neighbors=i, metric='euclidean') 
        scores = cross_val_score(kNN, X, y, scoring='accuracy', cv=kf)
        print(f'Accuracy (average): %.5f  %.5f | k = {i}' % (scores.mean(), scores.std()))
        print(f'Accuracy : %.5f | k = {i} {scores}')

        y_pred = cross_val_predict(kNN, X, y, cv=kf)
        print(confusion_matrix(y, y_pred))

print("\nPCA_cnn_VGG16_256_max:")
aplicar_KNN_with_kfold(X1, y1)

print("\nPCA_cnn_VGG16_256_avg:")
aplicar_KNN_with_kfold(X2, y2)

print("\nPCA_cnn_VGG19_128_max:")
aplicar_KNN_with_kfold(X3, y3)

print("\nPCA_cnn_VGG19_256_max:")
aplicar_KNN_with_kfold(X4, y4)

print("\nPCA_CNN_VGG19_128-avg:")
aplicar_KNN_with_kfold(X5, y5)

print("\nPCA_CNN_VGG19_256-avg:")
aplicar_KNN_with_kfold(X6, y6)
```
