```python
## Importando as libs necessárias
from google.colab import files
import os, io, time

# Criando o diretório inicial: /content/
os.chdir('/content/')
try:
  os.mkdir('Imagens', )
except:
  print('A pasta já existe.')
os.chdir('./Imagens')
os.listdir()
uploaded_images = files.upload()
```

```python
## Carregar as Libs necessarias
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.applications.vgg16 import VGG16, preprocess_input
from keras.applications.vgg19 import VGG19
import numpy as np

## Carregar os modelos de CNNs pre-treinadas (VGG16 e VGG19)
model_16 = VGG16(weights='imagenet', include_top=False, pooling='max')
model_19 = VGG19(weights='imagenet', include_top=False, pooling='max')
model_16_avg = VGG16(weights='imagenet', include_top=False, pooling='avg')
model_19_avg = VGG19(weights='imagenet', include_top=False, pooling='avg')

####### transformações ######
def aplicar_CNN(model_cnn, tamanho_image):
  cnn_features = []
  for filename in uploaded_images.keys():
    image = load_img(filename, target_size=(tamanho_image, tamanho_image))
    print(filename)
    x = img_to_array(image)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)

    features = model_cnn.predict(x)
    features_flatten_cnn = features.flatten()
    cnn_features.append(features_flatten_cnn)

  return cnn_features

```

```python
## Pegando as classes=de cada imagem lida
def extrair_classe():
  classe = []

  for filename in uploaded_images.keys():
    if 'Egyptian_Mau' in filename:
        classe.append('Egyptian_Mau')
    elif 'pomeranian' in filename:
        classe.append('Pomeranian')
    elif 'Ragdoll' in filename:
        classe.append('Ragdoll')
    elif 'shiba_inu' in filename:
        classe.append('Shiba_inu')

  return classe
```

```python
import pandas as pd

def salvar_cnn_csv(cnn_features, classe, nome_arquivo):
    # Criação do DataFrame
    data = {"cnn_features": cnn_features, "classe": classe}
    df = pd.DataFrame(data)

    # Expandindo os vetores HOG
    df2 = pd.DataFrame(df['cnn_features'].tolist())
    df2.columns = df2.columns.map(lambda x: f'cnn_feature_{x+1}')

    # Concatenando e organizando
    df = pd.concat([df.drop('cnn_features', axis=1), df2], axis=1)
    df['classe'] = df.pop('classe')

    # Salvando CSV
    df.to_csv(nome_arquivo, sep=',', index=False)
    print(f"Arquivo salvo como {nome_arquivo}")
```

```python
classe = extrair_classe()

cnn_features_VGG16_128_max = aplicar_CNN(model_16, 128)
cnn_features_VGG16_128_avg = aplicar_CNN(model_16_avg, 128)

cnn_features_VGG19_128_max = aplicar_CNN(model_19, 128)
cnn_features_VGG19_128_avg = aplicar_CNN(model_19_avg, 128)

cnn_features_VGG16_256_max = aplicar_CNN(model_16, 256)
cnn_features_VGG16_256_avg = aplicar_CNN(model_16_avg,256)

cnn_features_VGG19_256_max = aplicar_CNN(model_19, 256)
cnn_features_VGG19_256_avg = aplicar_CNN(model_19_avg, 256)

salvar_cnn_csv(cnn_features_VGG16_128_max, classe, 'cnn_VGG16_128_max.csv')
salvar_cnn_csv(cnn_features_VGG16_128_avg, classe, 'cnn_VGG16_128_avg.csv')
salvar_cnn_csv(cnn_features_VGG19_128_max, classe, 'cnn_VGG19_128_max.csv')
salvar_cnn_csv(cnn_features_VGG19_128_avg, classe, 'cnn_VGG19_128_avg.csv')
salvar_cnn_csv(cnn_features_VGG16_256_max, classe, 'cnn_VGG16_256_max.csv')
salvar_cnn_csv(cnn_features_VGG16_256_avg, classe, 'cnn_VGG16_256_avg.csv')
salvar_cnn_csv(cnn_features_VGG19_256_max, classe, 'cnn_VGG19_256_max.csv')
salvar_cnn_csv(cnn_features_VGG19_256_avg, classe, 'cnn_VGG19_256_avg.csv')
```
