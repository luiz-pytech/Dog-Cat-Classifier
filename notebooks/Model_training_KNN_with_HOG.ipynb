```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix
```

```python
## LENDO AS BASES HOGS E DEFININDO X E y PARA HOLDOUT ##
df1 = pd.read_csv('../Dados/HOG_Features/HOG_128_16x16.csv')
df2 = pd.read_csv('../Dados/Hog_Features/HOG_128_20x20.csv')
df3 = pd.read_csv('../Dados/Hog_Features/HOG_256_16x16.csv')
df4 = pd.read_csv('../Dados/Hog_Features/HOG_256_20x20.csv')

print(f"df1:{df1.shape} | df2:{df2.shape} | df3:{df3.shape} | df4:{df4.shape}")

def preparation_holdout(dataframe):
    df_features = dataframe.iloc[:, :-1]
    df_classe = dataframe['classe']

    return df_features, df_classe;

X1, y1 = preparation_holdout(df1)
X2, y2 = preparation_holdout(df2)
X3, y3 = preparation_holdout(df3)
X4, y4 = preparation_holdout(df4)
```

```python
## Treinamento e Teste: Holdout ##
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.3,shuffle=True, random_state=1)
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.3,shuffle=True, random_state=1)
X3_train, X3_test, y3_train, y3_test = train_test_split(X3, y3, test_size=0.3,shuffle=True, random_state=1)
X4_train, X4_test, y4_train, y4_test = train_test_split(X4, y4, test_size=0.3,shuffle=True, random_state=1)
```

```python
def aplicar_KNN_with_Holdout(X_train, X_test, y_train, y_test, k_range = 10):
    for i in range(1,  k_range + 1):
        kNN = KNeighborsClassifier(n_neighbors=i, metric='euclidean') 
        kNN.fit(X_train, y_train)
    
        y_pred = kNN.predict(X_test)
        print(f"Accuracy k = {i}:",metrics.accuracy_score(y_test, y_pred))
        print(confusion_matrix(y_test, y_pred))

## hog_128_16x16 ##
print("\nhog_128_16x16:")
aplicar_KNN_with_Holdout(X1_train, X1_test, y1_train, y1_test)

print("\n-------------------------------------------------------------------------------------------")

## hog_128_20x20 ##
print("\nhog_128_20x20:")
aplicar_KNN_with_Holdout(X2_train, X2_test, y2_train, y2_test)

print("\n-------------------------------------------------------------------------------------------")

## hog_256_16x16 ##
print("\nhog_256_16x16:")
aplicar_KNN_with_Holdout(X3_train, X3_test, y3_train, y3_test)

print("\n-------------------------------------------------------------------------------------------")

## hog_256_20x20 ##
print("\nhog_256_20x20:")
aplicar_KNN_with_Holdout(X4_train, X4_test, y4_train, y4_test)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
```

```python
def aplicar_KNN_with_kfold(X, y, k_range = 10, k_fold = 10):
    kf = KFold(n_splits=k_fold, random_state=1, shuffle=True)
    for i in range(1, k_range + 1):
         kNN = KNeighborsClassifier(n_neighbors=i, metric='euclidean') 
         scores = cross_val_score(kNN, X, y, scoring='accuracy', cv=kf)
         print(f'\nAccuracy: %.5f (%.5f) k = {i}' % (scores.mean(), scores.std()))
         print("Cross-val accuracies:", scores)

         y_pred = cross_val_predict(kNN, X, y, cv=kf)
         print(confusion_matrix(y, y_pred))

# Model Accuracy
print("\nhog_128_16x16:")
aplicar_KNN_with_kfold(X1, y1)

print("\n-------------------------------------------------------------------------------------------")

print("\nhog_128_20x20:")
aplicar_KNN_with_kfold(X2, y2)

print("\n-------------------------------------------------------------------------------------------")

print("\nhog_256_16x16:")
aplicar_KNN_with_kfold(X3, y3)

print("\n-------------------------------------------------------------------------------------------")

print("\nhog_256_20x20:")
aplicar_KNN_with_kfold(X4, y4)
```
